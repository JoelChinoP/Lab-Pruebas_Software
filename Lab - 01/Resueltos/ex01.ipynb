```python
#EJERCICIO 01
# Solicita base o
base = float(input("Ingresa la base del rectángulo (puede ser entero o decimal): "))

# Solicita altura
altura = float(input("Ingresa la altura del rectángulo (puede ser entero o decimal): "))

# Calcular el area
area = base * altura

# Muestra resultados
print(f"\nBase ingresada: {base}")
print(f"Altura ingresada: {altura}")
print(f"Área del rectángulo: {area}")
```

```python
#EJERCICIO 02
# CÓDIGO PRIMERA VERSIÓN
# Solicita cuántos números desea ingresar
cantidad = int(input("¿Cuántos números enteros deseas ingresar? "))

# Almacena los números
numeros = []

# Pide que ingrese los números
for i in range(cantidad):
    numero = int(input(f"Ingrese el número {i+1}: "))
    numeros.append(numero)

# Verifica si número es par o impar
print("\nResultado:")
for numero in numeros:
    if numero % 2 == 0:
        print(f"El número {numero} es par.")
    else:
        print(f"El número {numero} es impar.")
```

```python
#EJERCICIO 02
#CÓDIGO 2DA VERSIÓN
# Función para ingresar un número entero y validar que sea válido
def ingresar_numero_entero(mensaje):
    while True:
        entrada = input(mensaje)
        try:
            numero = int(entrada)
            return numero
        except ValueError:
            print("Entrada inválida. Por favor, ingresa un número entero válido.")

# Función para ingresar la cantidad de números y validar
def ingresar_cantidad():
    while True:
        entrada = input("¿Cuántos números enteros deseas ingresar? ")
        
        # Comprueba si la entrada es válida y un número entero positivo
        if entrada.strip() == "":  # Entrada vacía o solo espacios
            print("La cantidad no puede estar vacía. Por favor, ingresa un número válido.")
        else:
            try:
                cantidad = int(entrada)
                if cantidad > 0:
                    return cantidad
                else:
                    print("La cantidad debe ser un número positivo.")
            except ValueError:
                print("Entrada inválida. Por favor, ingresa un número entero válido.")

# Solicita cuántos números desea ingresar
cantidad = ingresar_cantidad()

# Almacena los números
numeros = []

# Pide al usuario que ingrese los números
for i in range(cantidad):
    while True:
        numero = ingresar_numero_entero(f"Ingrese el numero {i+1}: ")
        numeros.append(numero)
        break

# Verifica si cada número es par o impar
print("\nResultado:")
for numero in numeros:
    if numero % 2 == 0:
        print(f"El número {numero} es par.")
    else:
        print(f"El número {numero} es impar.")
```

```python
#EJERCICIO 03
# Saldo inicial
saldo = 1000.0

while True:
    print("\n=== Cajero Automático ===")
    print("1. Consultar Saldo")
    print("2. Depositar Dinero")
    print("3. Retirar Dinero")
    print("4. Salir")
    
    opcion = input("Selecciona una opción (1-4): ")
    
    if opcion == "1":
        print(f"\nTu saldo actual es: S/. {saldo:.2f}")
    
    elif opcion == "2":
        try:
            monto = float(input("Ingresa el monto a depositar: "))
            if monto > 0:
                saldo += monto
                print(f"Has depositado S/. {monto:.2f}. Saldo actualizado: S/. {saldo:.2f}")
            else:
                print("El monto debe ser positivo.")
        except ValueError:
            print("Por favor, ingresa un número válido.")

    elif opcion == "3":
        try:
            monto = float(input("Ingresa el monto a retirar: "))
            if monto <= 0:
                print("El monto debe ser positivo.")
            elif monto > saldo:
                print("Fondos insuficientes.")
            else:
                saldo -= monto
                print(f"Has retirado S/. {monto:.2f}. Saldo restante: S/. {saldo:.2f}")
        except ValueError:
            print("Por favor, ingresa un numero válido.")

    elif opcion == "4":
        print("Gracias por usar el cajero automático. ¡Hasta pronto!")
        break

    else:
        print("Opción no valida. Intenta nuevamente.")
```
